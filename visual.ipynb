```python
import matplotlib.pyplot as plt
import seaborn as sns
from process import *
```

```python
data = Process_Data(DIR, True)
```

```python
data.raw_data['Aotizhongxin']
```

```python
df = dict()
for station in data.stations:
    df[station] = data.process_data[station]
df['Aotizhongxin']
```

```python
df_arr = list()
for i, station in enumerate(df.keys()):
    df_arr.append(df[station])
```

```python
def draw(arg):
    if arg in df_arr[0].columns:
        year = df_arr[0].index.year.unique()
        plot_rows = len(year)
        size = 10, 25
        plt.figure(figsize=size)
        for i in range(plot_rows):
            plt.subplot(plot_rows, 3, i+1)
            pm_2_5_data_idx = df_arr[0][arg][f'{year[i]}-01-01':f'{year[i]}-12-31']
            month = pm_2_5_data_idx.index.month.unique()
            val = pm_2_5_data_idx.groupby(pm_2_5_data_idx.index.month).mean()
            plt.rc('axes', labelsize=7.5)
            plt.bar(month, val)
            plt.plot(month, val, color='r')
            plt.xlabel('Month', fontsize=9)
            plt.title(f'{year[i]}', fontsize=13)
```

```python
draw('PM2.5')
```

```python
# Temperature by year
draw('TEMP')
```

```python
# Pressure by year
draw('PRES')
```

```python
# Dew point temperature (DEWP)
draw('DEWP')
```

```python
# Rain by year
draw('RAIN')
```

```python
# Wind speed by year
draw('WSPM')
```

```python
# Các giá trị lớn nhất và nhỏ nhất tại Aotizhongxin
max_vals_0 = data.max_vals['Aotizhongxin']
min_vals_0 = data.min_vals['Aotizhongxin']

# Phân tích 'PM2.5', 'TEMP', 'PRES', 'DEWP', 'RAIN','wd', 'WSPM'
pm_2_5_meteo_cols = ['PM2.5', 'TEMP', 'PRES', 'DEWP', 'RAIN','wd', 'WSPM']
plot_rows = len(pm_2_5_meteo_cols)
i = 1
plt.figure(figsize=(25,10))
for col in pm_2_5_meteo_cols:
    plt.subplot(plot_rows, 1 , i)
    i += 1
    plt.plot(np.arange(len(df_arr[0][col][0:1000])), df_arr[0][col][0:1000], 'g', linewidth = 2)
    plt.title(col+'\nMax:{}\nMin:{}'.format(max_vals_0[col], min_vals_0[col]), x=0.025,y=0.06)
```

```python
train_cols = pm_2_5_meteo_cols
df_arr[0][train_cols]
fig, ax = plt.subplots(figsize=(25,10))
corr_data = df_arr[0][train_cols].corr(method = 'pearson')
corr_data = corr_data.fillna(value = 0)
ones_corr = np.ones_like(corr_data, dtype=bool)
mask = np.triu(ones_corr)
sns.heatmap(data=corr_data, mask=mask, annot=True, cmap = 'Greens', linewidths=.5)
```
